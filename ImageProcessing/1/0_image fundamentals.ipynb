### Loading and showing an image with Scikit-image

##### Documentation: https://scikit-image.org/docs/stable/api/api.html

##### Built-in images

```python
image = skimage.data.coins()
skimage.io.imshow(image)
skimage.io.show()
```

#### Loading from disk

```python
lena_im = skimage.io.imread("images/lenna-RGB.tif")
skimage.io.imshow(lena_im)
```

```python
print(type(lena_im))
print("Shape of the image is {} and the type {}".format(lena_im.shape, lena_im.dtype))
```

We see that Scikit-image loads images as data type numpy.ndarray. This is the common data type for matrices and higher dimensional arrays in Python as there is no built-in type for these.

#### Plotting with Matplotlib
##### Matplotlib is a popular plotting library. It let't you plot both images and scatter plots, curves, bars, etc. The skimage.io.imshow() function above actually is a only wrapper on top of matplotlib.
##### Documentation: https://matplotlib.org/stable/api/index.html

```python
plt.imshow(lena_im)
```

Plotting only the red color channel:

```python
plt.imshow(lena_im[:, :, 0], cmap=cm.gray)
```

### Accessing pixels

It is important to understand which type of data do you obtain once the image is loaded. Subsequent algorithms will require a certain type (*ie.*, float, unsigned int) included in a certain data range.

```python
lena_im
```

You can see that this is an array of arrays of arrays (R x C x 3). The data type of each element is uint8 (unsigned 8 bit integer), which is very common for digital images.
We can easily modify rectangular subregions of the array.

```python
lena_im[100:150, 100:150, :] = 255
plt.imshow(lena_im, cmap=cm.gray)
```

### Histograms
Now lets see the distribition of the intensity values within a image.

```python
bins, bins_c = skimage.exposure.histogram(lena_im[:, :, 1])
plt.plot(bins_c, bins)
```

```python
bins, bins_e = np.histogram(lena_im[:, :, 1], 255)
plt.plot(bins_e[1:], bins)
```
