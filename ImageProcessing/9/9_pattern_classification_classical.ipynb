```python
import numpy as np
import matplotlib.pyplot as plt
import skimage
from random import randint
```

```python
def plotImage(image, title):
    if(len(image.shape) == 2):
        plt.imshow(image, cmap='gray')
    else:
        plt.imshow(image)
    plt.title(title)
    plt.axis('off')
    plt.show()
```
